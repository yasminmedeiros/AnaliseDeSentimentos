### 3.2 Analisando os 20 personagens que mais apresentam falas

#### Armazenando na variável personagens os 50 personagens dos diálogos que possuem mais falas

```python
typeofpersonagens = nltk.FreqDist(simpsonsData['raw_character_text'])
personagens = [x[0] for x in typeofpersonagens.most_common(41) if pd.notna(x[0])]
personagens
```

#### Criando um dicionário, separando as falas de cada personagem

```python
dictSimpsonsPersonagens = {}
for x in personagens:
    dictSimpsonsPersonagens[x] = [] 
```

```python
for y in dictSimpsonsPersonagens.keys():
    for x in range(len(simpsonsData)):
        if simpsonsData.loc[x,'raw_character_text'] == y:
            a = dictSimpsonsPersonagens[y]
            a.append(simpsonsData.loc[x,'spoken_words'])
            dictSimpsonsPersonagens[y] = a
      
```

### 3.1.1 Banco de Dados 1
#### Ferramenta de Análise 1: SentimentAnalyzer é uma ferramenta para implementar e facilitar tarefas de análise de sentimento usando recursos e classificadores do NLTK, especialmente para fins de ensino e demonstração.

```python
df = pd.DataFrame()
sid = SentimentIntensityAnalyzer()
i=0
for x in [x for x in dictSimpsonsPersonagens.keys() if type(x)==str]:
    for y in [z for z in dictSimpsonsPersonagens[x] if type(z)==str]:
        df.loc[i,'character'] = x
        df.loc[i,'text'] = y
        print(df.loc[i,'text'])
        ss = sid.polarity_scores(y)
        for k in sorted(ss):
            df.loc[i,k] = ss[k]
            print('{0}: {1}, '.format(k, ss[k]), end='')
        i+=1
```

#### Identificando qual o sentimento mais predominante em cada fala de cada personagem.

```python
for x in range(len(df)):
    if(df.loc[x,'neg']>df.loc[x,'pos'] and df.loc[x,'neg']>df.loc[x,'neu']):
        df.loc[x,'major'] = 'neg'
    elif (df.loc[x,'pos']>df.loc[x,'neg'] and df.loc[x,'pos']>df.loc[x,'neu']):
        df.loc[x,'major'] = 'pos'
    else:
        df.loc[x,'major'] = 'neu'
```

```python
def most_feeling_per_character(df, character):
    a = df[df['character']==character]
    neg,pos = 0,0
    for x in a['major']:
        if x =="neg":
            neg+=1
        elif x == "pos":
            pos+=1
    
    print(character)
    
    if (pos>neg):
        #print("pos major than neg",pos,neg)
        return "pos"
    elif(neg>pos):
        #print("neg major than pos",pos,neg)
        return "neg"
            
    else:
        print("neg equal pos",pos,neg)
```

```python
character = pd.DataFrame(columns=['character','major'])
i=0
for x in [x for x in personagens if type(x)==str]:
    character.loc[i,'character']=x
    character.loc[i,'major'] = most_feeling_per_character(df,x)
    character.loc[i,'neu'] = pd.DataFrame([df.loc[z,'neu'] for z in range(len(df)) if df.loc[z,'character'] == x]).mean()[0]
    character.loc[i,'pos'] = pd.DataFrame([df.loc[z,'pos'] for z in range(len(df)) if df.loc[z,'character'] ==  x]).mean()[0]
    character.loc[i,'neg'] = pd.DataFrame([df.loc[z,'neg'] for z in range(len(df)) if df.loc[z,'character'] ==  x]).mean()[0]
    if character.loc[i,'neg'] > character.loc[i,'pos']:
        character.loc[i,'major_sentiment'] = 'neg'
    else: 
        character.loc[i,'major_sentiment'] = 'pos'
    i+=1
```
